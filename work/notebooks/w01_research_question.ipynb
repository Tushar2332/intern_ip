## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

The Unit of Analysis: A single published webpage (represented by content_id), analyzed over a specific time window.
The Decision & Action: A content manager must decide which declining or underperforming pages to assign to a writing team. The explicit action is a human reviewer rewriting, expanding, or consolidating the recommended page.
The Cost of a Wrong Call: A false positive wastes a writer's expensive time on a page that cannot recover (or was just experiencing seasonal fluctuation). A false negative means leaving fixable traffic and revenue on the table while competitors capture it.
Why ML? Static rules (e.g., "review all pages that dropped 10% in traffic") are too brittle and create massive, unprioritized backlogs. Machine learning can weigh multiple interacting variables—like search volume, competition, current position tier, and historical engagement—to predict risk and opportunity far more accurately than a human writing nested if/else statements.
